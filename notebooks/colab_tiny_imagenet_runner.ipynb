```python
DATASET = 'tiny_imagenet'
IMAGE_SIZE = 64
BATCH_SIZE = 128
AUGMENT = 'basic'
VAL_SIZE = 10000
EPOCHS_TEACHER = 6
EPOCHS_STUDENT = 8
MAX_TRAIN_BATCHES = 0
MAX_VAL_BATCHES = 0
```

```python
TEACHER_OUT = OUTPUT_ROOT / 'teacher_resnet50_quality'
B_TINYCNN_OUT = OUTPUT_ROOT / 'baseline_tiny_cnn'
D_TINYCNN_OUT = OUTPUT_ROOT / 'distill_tiny_cnn'
B_MBV3_OUT = OUTPUT_ROOT / 'baseline_mobilenetv3_small'
D_MBV3_OUT = OUTPUT_ROOT / 'distill_mobilenetv3_small'
CMP_STUDENTS_OUT = OUTPUT_ROOT / 'compare_students'
CMP_VARIANTS_TINYCNN_OUT = OUTPUT_ROOT / 'compare_variants_tinycnn'
```

```python
# 1) Train strong teacher on Tiny ImageNet
!python -m src.train_teacher \
  --dataset {DATASET} \
  --data-dir {DATA_DIR} \
  --teacher-model resnet50 \
  --teacher-pretrained \
  --finetune-mode full \
  --augment {AUGMENT} \
  --image-size $IMAGE_SIZE \
  --channels-last \
  --batch-size $BATCH_SIZE \
  --val-size $VAL_SIZE \
  --epochs $EPOCHS_TEACHER \
  --max-train-batches $MAX_TRAIN_BATCHES \
  --max-val-batches $MAX_VAL_BATCHES \
  --output-dir {TEACHER_OUT}
```

```python
# 2) Baseline + distillation for weaker student: tiny_cnn
!python -m src.train_baseline \
  --dataset {DATASET} \
  --data-dir {DATA_DIR} \
  --student tiny_cnn \
  --augment {AUGMENT} \
  --image-size $IMAGE_SIZE \
  --channels-last \
  --batch-size $BATCH_SIZE \
  --val-size $VAL_SIZE \
  --epochs $EPOCHS_STUDENT \
  --max-train-batches $MAX_TRAIN_BATCHES \
  --max-val-batches $MAX_VAL_BATCHES \
  --output-dir {B_TINYCNN_OUT}

!python -m src.train_distill \
  --dataset {DATASET} \
  --data-dir {DATA_DIR} \
  --student tiny_cnn \
  --teacher resnet50 \
  --teacher-checkpoint {TEACHER_OUT / 'teacher_best.pt'} \
  --augment {AUGMENT} \
  --embed-loss mse \
  --cls-weight 1.0 \
  --embed-weight 1.0 \
  --logit-weight 0.0 \
  --image-size $IMAGE_SIZE \
  --channels-last \
  --batch-size $BATCH_SIZE \
  --val-size $VAL_SIZE \
  --epochs $EPOCHS_STUDENT \
  --max-train-batches $MAX_TRAIN_BATCHES \
  --max-val-batches $MAX_VAL_BATCHES \
  --output-dir {D_TINYCNN_OUT}
```

```python
# 3) Baseline + distillation for MobileNetV3-small
!python -m src.train_baseline \
  --dataset {DATASET} \
  --data-dir {DATA_DIR} \
  --student mobilenetv3_small \
  --augment {AUGMENT} \
  --image-size $IMAGE_SIZE \
  --channels-last \
  --batch-size $BATCH_SIZE \
  --val-size $VAL_SIZE \
  --epochs $EPOCHS_STUDENT \
  --max-train-batches $MAX_TRAIN_BATCHES \
  --max-val-batches $MAX_VAL_BATCHES \
  --output-dir {B_MBV3_OUT}

!python -m src.train_distill \
  --dataset {DATASET} \
  --data-dir {DATA_DIR} \
  --student mobilenetv3_small \
  --teacher resnet50 \
  --teacher-checkpoint {TEACHER_OUT / 'teacher_best.pt'} \
  --augment {AUGMENT} \
  --embed-loss mse \
  --cls-weight 1.0 \
  --embed-weight 1.0 \
  --logit-weight 0.0 \
  --image-size $IMAGE_SIZE \
  --channels-last \
  --batch-size $BATCH_SIZE \
  --val-size $VAL_SIZE \
  --epochs $EPOCHS_STUDENT \
  --max-train-batches $MAX_TRAIN_BATCHES \
  --max-val-batches $MAX_VAL_BATCHES \
  --output-dir {D_MBV3_OUT}
```

```python
# 4) Compare students
!python -m src.compare_students \
  --run tiny_cnn:{B_TINYCNN_OUT / 'metrics.json'}:{D_TINYCNN_OUT / 'metrics.json'} \
  --run mobilenetv3_small:{B_MBV3_OUT / 'metrics.json'}:{D_MBV3_OUT / 'metrics.json'} \
  --output-dir {CMP_STUDENTS_OUT}
```

```python
# 5) Extra ablation on tiny_cnn: embed-only vs embed+cls vs logit+cls
D_TINYCNN_EMBED_ONLY = OUTPUT_ROOT / 'distill_tinycnn_embed_only'
D_TINYCNN_LOGIT_CLS = OUTPUT_ROOT / 'distill_tinycnn_logit_cls'

!python -m src.train_distill \
  --dataset {DATASET} \
  --data-dir {DATA_DIR} \
  --student tiny_cnn \
  --teacher resnet50 \
  --teacher-checkpoint {TEACHER_OUT / 'teacher_best.pt'} \
  --augment {AUGMENT} \
  --embed-loss mse \
  --cls-weight 0.0 \
  --embed-weight 1.0 \
  --logit-weight 0.0 \
  --image-size $IMAGE_SIZE \
  --channels-last \
  --batch-size $BATCH_SIZE \
  --val-size $VAL_SIZE \
  --epochs $EPOCHS_STUDENT \
  --max-train-batches $MAX_TRAIN_BATCHES \
  --max-val-batches $MAX_VAL_BATCHES \
  --output-dir {D_TINYCNN_EMBED_ONLY}

!python -m src.train_distill \
  --dataset {DATASET} \
  --data-dir {DATA_DIR} \
  --student tiny_cnn \
  --teacher resnet50 \
  --teacher-checkpoint {TEACHER_OUT / 'teacher_best.pt'} \
  --augment {AUGMENT} \
  --embed-loss mse \
  --cls-weight 1.0 \
  --embed-weight 0.0 \
  --logit-weight 1.0 \
  --temperature 2.0 \
  --image-size $IMAGE_SIZE \
  --channels-last \
  --batch-size $BATCH_SIZE \
  --val-size $VAL_SIZE \
  --epochs $EPOCHS_STUDENT \
  --max-train-batches $MAX_TRAIN_BATCHES \
  --max-val-batches $MAX_VAL_BATCHES \
  --output-dir {D_TINYCNN_LOGIT_CLS}

!python -m src.compare_variants \
  --run baseline_cls:{B_TINYCNN_OUT / 'metrics.json'} \
  --run embed_plus_cls:{D_TINYCNN_OUT / 'metrics.json'} \
  --run embed_only:{D_TINYCNN_EMBED_ONLY / 'metrics.json'} \
  --run logit_plus_cls:{D_TINYCNN_LOGIT_CLS / 'metrics.json'} \
  --title 'Tiny ImageNet: tiny_cnn ablations' \
  --output-dir {CMP_VARIANTS_TINYCNN_OUT}
```
